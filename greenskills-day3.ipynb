```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Solar":[1200,1600,1300],
    "Wind":[3400,3600,3200],
    "Hydropower":[2900,3100,2800],
    "Biomass": [2500,2700,2400]
}

df = pd.DataFrame(data)
sns.pairplot(df)
plt.show()
```

```python
correlation_matrix = df.corr()
plt.title("Correlation between Energy Sources")
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm', linewidth = 0.5)
```

```python
sns.boxplot(data = df)
plt.title("Distribution")
plt.show()
```

```python
sns.violinplot(data = df)
plt.title("Violin")
plt.show()
```

```python
energy_values = [100, 200, 300, 400, 500]
carbon_emissions = [10, 20, 25, 40, 50]

df_reg = pd.DataFrame({
    'Energy Consumption': energy_values,
    'Carbon Emissions': carbon_emissions
})

sns.regplot(x = 'Energy Consumption', y = 'Carbon Emissions', data = df_reg)
plt.show()
```

3D Graph plots

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

ax = plt.axes(projection = '3d')

z = np.linspace(0,1,100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("IRIS.csv")
df = pd.DataFrame(data)
df.head()
```

```python
sns.pairplot(df, hue = 'species')
plt.title("Correlation between flowers");
plt.show()
```

```python
qqqqq = df[["sepal_length","sepal_width","petal_length","petal_width"]]
correlation_matrix = qqqqq.corr()
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm', linewidth = 0.5)
plt.title("ok")
plt.show()
```

```python
sns.boxplot(data = df)
plt.title("Not ok")
plt.show()
```
